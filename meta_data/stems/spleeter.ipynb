```python
import os
import subprocess
import sys

# Set paths
input_file = "/content/rhythm_machine/Rhythm_Machine_segment_2.wav"
output_dir = "/content/processed_audio"

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Try multiple invocation methods
try:
    # Method 1: Direct command with explicit file argument
    print("Method 1: Direct Command")
    result1 = subprocess.run([
        'spleeter', 'separate',
        input_file,
        '-p', 'spleeter:2stems',
        '-o', output_dir
    ],
    capture_output=True,
    text=True,
    check=True)

    print(result1.stdout)
    print(f"Processing complete. Output directory: {output_dir}")

except subprocess.CalledProcessError as e:
    print("\nMethod 1 Error:")
    print("Return Code:", e.returncode)
    print("STDOUT:", e.stdout)
    print("STDERR:", e.stderr)

    # Method 2: Python module invocation
    try:
        print("\nMethod 2: Python Module")
        result2 = subprocess.run([
            sys.executable, '-m', 'spleeter', 'separate',
            input_file,
            '-p', 'spleeter:2stems',
            '-o', output_dir
        ],
        capture_output=True,
        text=True,
        check=True)

        print(result2.stdout)
        print(f"Processing complete. Output directory: {output_dir}")

    except subprocess.CalledProcessError as e:
        print("\nMethod 2 Error:")
        print("Return Code:", e.returncode)
        print("STDOUT:", e.stdout)
        print("STDERR:", e.stderr)

except Exception as e:
    print("\nUnexpected Error:")
    print(type(e))
    print(str(e))
```

```python
import os
import subprocess
import sys

# Set input and output directories
input_dir = "/content/Chromatic"  # Directory containing audio files
output_dir = "/content/processed_audio"  # Directory to save separated stems

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Get list of audio files to process
# Add or modify extensions as needed
audio_extensions = ['.wav', '.mp3', '.flac', '.ogg']
audio_files = [
    os.path.join(input_dir, f)
    for f in os.listdir(input_dir)
    if os.path.isfile(os.path.join(input_dir, f))
    and os.path.splitext(f)[1].lower() in audio_extensions
]

# Process each audio file
processed_files = []
errors = []

for input_file in audio_files:
    try:
        # Prepare output subdirectory for this file
        filename = os.path.splitext(os.path.basename(input_file))[0]
        file_output_dir = os.path.join(output_dir, filename)

        # Run Spleeter
        result = subprocess.run([
            'spleeter', 'separate',
            input_file,
            '-p', 'spleeter:2stems',
            '-o', output_dir
        ],
        capture_output=True,
        text=True,
        check=True)

        processed_files.append(input_file)
        print(f"Processed: {input_file}")
        print(result.stdout)

    except subprocess.CalledProcessError as e:
        print(f"\nError processing {input_file}:")
        print("Return Code:", e.returncode)
        print("STDERR:", e.stderr)
        errors.append((input_file, e.stderr))

# Summary
print("\n--- Processing Summary ---")
print(f"Total files found: {len(audio_files)}")
print(f"Successfully processed: {len(processed_files)}")
print(f"Errors: {len(errors)}")

# Print error details if any
if errors:
    print("\nError Details:")
    for file, error in errors:
        print(f"File: {file}")
        print(f"Error: {error}\n")

print(f"\nOutput directory: {output_dir}")
```
